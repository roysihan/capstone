```python
import requests
import json
from IPython.display import Audio, display

# Masukkan API key Anda di sini
API_KEY = '7681ffea73c203defd3af8e555598972'
VOICE_ID = 'EXAVITQu4vr4xnSDxMaL'  # Ganti dengan ID suara yang ingin Anda gunakan
API_URL = f'https://api.elevenlabs.io/v1/text-to-speech/{VOICE_ID}'

# Fungsi untuk mengubah teks menjadi ucapan
def text_to_speech(text, voice_id):
    headers = {
        'Content-Type': 'application/json',
        'xi-api-key': API_KEY
    }
    data = {
        'text': text,
        'voice_settings': {
            'stability': 0.75,
            'similarity_boost': 0.75
        }
    }

    response = requests.post(API_URL, headers=headers, data=json.dumps(data))

    if response.status_code == 200:
        audio_content = response.content
        with open('output_voice.mp3', 'wb') as audio_file:
            audio_file.write(audio_content)
        print("Voice changed successfully. Output saved as 'output_voice.mp3'")
        display(Audio('output_voice.mp3', autoplay=True))
    else:
        print("Error:", response.status_code, response.text)

# Contoh teks dalam bahasa Indonesia
text = "Halo, ini adalah contoh konversi teks ke ucapan menggunakan ElevenLabs API."

# Panggil fungsi untuk mengubah teks menjadi ucapan
text_to_speech(text, VOICE_ID)
```
